```python
import pandas as pd
```

```python
df = pd.read_csv('./files/Salaries.csv', low_memory=False)
df.head
```

```python
df.dropna(subset=['TotalPay'])
```

```python
df.shape
```

```python
df = df[df['TotalPay'] != 0.00]
df.shape
```

```python
df
```

```python
df_low_slary = df[df['TotalPay'] < df['TotalPay'].mean()]
df_low_slary.shape 
```

```python
df_police_department = df[df['JobTitle'].str.contains('Police', na=False)]
df_police_department.shape
```

```python
mean_salary_in_police = df_police_department['TotalPay'].mean()
max_salary_in_police = df_police_department['TotalPay'].max()
min_salary_in_police = df_police_department['TotalPay'].min()
print(mean_salary_in_police)
print(min_salary_in_police)
```
