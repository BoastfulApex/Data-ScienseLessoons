```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 11, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1, 3, 1)
plt.plot(y, x, 'b')

plt.subplot(1, 3, 2)
plt.plot(x, y, 'r')

plt.subplot(1, 3, 3)
plt.plot(y, x, 'y')
```
